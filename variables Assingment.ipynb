```python
# Area & Perimeter of Rectangle
# Do variables length aur width lo.
# Area aur perimeter dono calculate karo.
# Output kuch aisa ho:
lenght = 10
weidth = 5
area = lenght * weidth
perimeter = 2 * (lenght + weidth)
print("perimeter:", perimeter)
print("area:", area) 
```

```python
# swap Two Variables Without Using a Third Variable.
# Example: a = 5, b = 10
# Swap karne ke baad a me 10 aur b me 5 aana chahiye.
a = 10
b = 15
a,b = b,a
print(a)
print(b)
```

```python
# Total Marks & Percentage
# 5 subjects ke marks lo (variables me store karo).
# Total aur percentage calculate karo (Assume each subject is out of 100).
english = 60
urdu = 65
math = 89
sindhi = 55
accouning = 94
total = english + urdu + math + sindhi + accouning
percentage = (total/ 500) *100
print ("Total",total)
print("Percentage",percentage)
```

```python
# Simple Interest Calculator
# Variables lo: P (principal), R (rate), T (time in years).
# Formula: SI = (P * R * T) / 100
# Output me print karo "Simple Interest is: X"
principal = 10000
Rate = 5
Time = 2
Formula = principal * Rate * Time / 100
print(Formula)
```

```python
# Temperature Converter
# Variable temp_f me Fahrenheit temperature rakho.
# Usko Celsius me convert karo using:
# C = (F - 32) * 5/9
# Output me dono show karo.
temp_f = 98.6
C = (temp_f - 32) * 5/9
print("Temperature Fahrenheit", temp_f)
print("Celsius", C)
```

```python
# Days to Years, Weeks, Days
# Variable days me total din rakho (jaise 500).
# Calculate karo:
# kitne years,
# kitne weeks,
# aur remaining days bache.
# (Hint: 1 year = 365 days, 1 week = 7 days)
D = 500
print(D//365)
print(D//7)
print(D%7)
```
