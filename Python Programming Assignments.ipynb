````python
### Assignment 2: Set Operations
# **Task:**
# Create two sets:
#`A = {1, 2, 3, 4}`
#`B = {3, 4, 5, 6}`
# Print their:
# *   Union (all unique elements from both sets)
# *   Intersection (elements common to both sets)
# *   Difference (elements in A but not in B)
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}
print("Union", A.union(B)) # Dono sets ke sab unique elements
print("Intersection", A.intersection(B)) # Sirf common elements
print("Difference", A.difference(B)) # Jo A mein hain lekin B mein nahi
````

````python
### Assignment 3: Loop Through Dictionary
#**Task:**
# Create a dictionary of 3 countries with their capitals (e.g., Pakistan: Islamabad, Turkey: Ankara, Japan: Tokyo).
# Loop through the dictionary and print each country and its capital in the format: `Country ‚Üí Capital`.
countries = {
    "Pakistan": "Islamabad",
    "Turkey": "Ankara",
    "Japan": "Tokyo"
}
for country, capital in countries.items():
    for country, capital in countries.items():
        print(f"{country} ‚Üí {capital}")
````

````python
### Assignment 4: Dictionary of Prices
# **Task:**
# Create a dictionary of 3 products and their prices (e.g., `{"apple": 1.5, "banana": 0.75, "orange": 1.2}`).
# Ask the user which product's price they want to see and display it.
# Create a dictionary of products and their prices
prices = {
    "apple": 1.5,
    "banana": 0.75,
    "orange": 1.2
}
# Ask the user for a product name
product = input("Enter the product name (apple, banana, orange): ")

# Display the price if the product exists
if product in prices:
    print(f"The price of {product} is ${prices[product]}")
else:
    print("Sorry, that product is not available.")
````
